扩散模型（Diffusion Model）[参考博客1](https://zhuanlan.zhihu.com/p/586936791) [参考博客2](https://zhuanlan.zhihu.com/p/565901160) 

扩散模型是通过将一个原始图片逐渐加入噪声，变成一个高斯噪声，然后逐步还原的做法。理解Diffusion的关键在于：

1、理解Diffusion的正向过程：需要注意，$P_{\theta}(x_{t}|x_{t-1})$是一个正态分布，其是对于$x_{t-1}$的一个正态采样；$P_{\theta}(x_{t}；x_0)$也是一个正态分布，其均值方差都是关于$x_0$的一个函数。

2、理解Diffusion的逆向过程：需要注意，逆向过程实际上是一个后验概率的分布，即$P_{\theta}(x_{t-1}|x_{t},x_0)$，按照数学原理，当$\beta$很小的时候，这也是一个正态分布，其均值和方差是$x_0，x_t$的一个函数。但是在推理阶段，我们是不知道$x_0$长什么样子的，因此我们利用了变分推断，希望可以使用一个分布$Q_{\theta}{(x_{t-1}|x_{t})}$来估计$P_{\theta}(x_{t-1}|x_{t},x_0)$，使得反向过程与$x_0$无关。注意，这两个分布都是正态分布，同时$P$分布的方差是固定的。因此我们希望优化着这两个分布的均值。对于均值的推导请参考[参考博客2](https://zhuanlan.zhihu.com/p/565901160)中的(14)~（21）。相当于预测一个的高斯噪声。

3、如何理解为什么预测的是高斯噪声？那么为什么不直接推理的时候使用这个采样一个高斯噪声，来得到$P_{\theta}(x_{t}|x_{t-1})$的分布？这是因为，在反向的过程中，高斯噪声$\epsilon$已经是一个定值了，不再是一个随机的值了。尽管期望上来看，他们是相同的，但是这样方差很大。因为，模型需要预测先采样一个高斯噪声，这样的高斯噪声下图片会变成什么样，然后再进行还原。相当于正向过程中采样的方差，组成了不同的进化路线，那我们方向的过程也需要逆着这个过程。

4、变分推断中的一个关键地方用到了$x_t$与$x_0$的关系，这里期望上来说变成了一个随机采样的高斯噪声，但是其实实际上是多个高斯噪声的不断叠加。因此在越深的层$x_t$的递推公式和$x_0$的方差越小。也就是[参考博客1](https://zhuanlan.zhihu.com/p/586936791)中的下列推导：![image.png](picture/diffusion1.png)。
